```python
#find nearby resutrants based on coordinates using the google maps API, I selected a 1/2 a km radius
coordinates = [['39.96690336626812', '-86.0089268190694'],['39.84895874219581','-86.14106948656577'], ['39.95155807619608', '-86.2608697018842'], ['39.77257944720804', '-86.15307701538342'], ['39.75953827532731', '-86.1580467018889'] ]
places = ["46038", "46220", "46077",  "46204", "46225"]
dictionary = dict(zip(places, coordinates))
print(dictionary)
keywords = ['restaurant']
radius = '500' #1/2 km
api_key  = 'AIzaSyD0RU0ZJM7dgBkzNmU0pg6K3UocM9cO90s'
final_data = []
for coordinate in coordinates:
  url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+coordinate[0]+'%2C'+coordinate[1]+'&radius='+str(radius)+'&keyword='+str(keywords[0])+'&key='+str(api_key)
  respon = requests.get(url)
  jj = json.loads(respon.text)
  results = jj['results']
  for result in results:
    name = result['name']
    place_id = result ['place_id']
    types = result['types']
    vicinity = result['vicinity']
    data = [coordinate[0]+" " +coordinate[1], name, types, vicinity]
    final_data.append(data)
labels = ['Coordinate', 'Place Name', 'Types', 'Vicinity']
food_data = pd.DataFrame.from_records(final_data, columns=labels)
print(food_data)
#export_dataframe_1_medium.to_csv('export_dataframe_1_medium.csv')
```

```python
#find nearby events based on zipcode of workspace, radius = 10

codes = ['46220', '46077', '46038', '46204', '46225']
keywords = ['Concerts', 'Arts & Theater', 'Family', 'Film', 'Sport']
events_data = []
for code in codes:
  for keyword in keywords:
    url = "https://app.ticketmaster.com/discovery/v2/events.json?postalCode=" + code + "&keyword=" + keyword + "&radius=10&unit=miles&apikey=vHERWiKypGvh9aaFmHs9CAOyOALpBKdw"

    # url = "https://app.ticketmaster.com/discovery/v2/events.json?+"&keyword="+keyword+"&dmaId=527&apikey=vHERWiKypGvh9aaFmHs9CAOyOALpBKdw
    respon = requests.get(url)
    jj = json.loads(respon.text)
    try:
      for event in jj['_embedded']['events']:
         name = event['name']
         genre =event['classifications'][0]['genre']['name']
         start = event['sales']['public']['startDateTime']
         end = event['sales']['public']['endDateTime']
         data = [code, name, genre, start, end]
         events_data.append(data)
    except:
      continue
    time.sleep(1) #prevent api from timing out
labels = ["Zip_Code", "Name", "Genre", "Start", "End"]
events_df = pd.DataFrame.from_records(events_data, columns=labels)
events_df["Name"] = events_df["Name"].drop_duplicates()
unique_events_df = events_df.dropna()
print(unique_events_df)



```
